```python
# music_instrument_agent.py

import os
from dotenv import load_dotenv
from langchain.agents import initialize_agent, Tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents.agent_types import AgentType

# Load Gemini API key
load_dotenv()
GOOGLE_API_KEY = "AIzaSyCUIHn7gtq7La_2ubYeBUNBsHzVtfHesbw"

# Setup Gemini model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7,
    google_api_key=GOOGLE_API_KEY
)

# Define custom tool - Music Knowledge Base (LLM based)
def ask_music_expert(query: str) -> str:
    prompt = f"You are a world-class musical instrument expert tutor. Answer this in expert detail:\n\n{query}"
    return llm.invoke(prompt)

# Wrap it as a LangChain Tool
music_tool = Tool(
    name="MusicInstrumentExpert",
    func=ask_music_expert,
    description="Gives expert-level answers on musical instruments (history, construction, brands, playing techniques, tuning, etc.)"
)

# Initialize the Agent
agent = initialize_agent(
    tools=[music_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Ask something
if __name__ == "__main__":
    while True:
        query = input("\nAsk about a musical instrument: ")
        if query.lower() in ['exit', 'quit']:
            break
        response = agent.run(query)
        print("\nðŸŽ» Answer:", response)
```
